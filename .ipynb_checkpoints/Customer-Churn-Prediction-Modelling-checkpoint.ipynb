## 3. Data cleaning

```python
#replacing 'No internet service' to No

NoInternet = [ 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                'TechSupport','StreamingTV', 'StreamingMovies']
for i in NoInternet: 
    dataset[i]  = dataset[i].replace({'No internet service' : 'No'})
    
#replacing 'No phone service' to No

dataset.MultipleLines = dataset.MultipleLines.replace({'No phone service':'No'})
```

```python
for i in NoInternet:
    print(dataset[i].value_counts())
```

```python
for i in dataset.TotalCharges.values:
    try:
        float(i)
    except Exception as e:
        print(e)
```

```python
dataset = dataset[dataset['TotalCharges'] != ' '] 
```

```python
# Now changing the dtype from object to float
dataset['TotalCharges'] = dataset['TotalCharges'].astype('float64')
```

```python
dataset.describe()
```

### 5. Feature engineering

```python
dataset.columns
```

```python
ynmap = {'Yes':1,'No':0}
```

```python
dataset['gender'] = dataset['gender'].replace({'Male':1,'Female':0})
```

```python
dataset.gender.value_counts()
```

```python
for i in NoInternet:
    dataset[i] = dataset[i].replace({'Yes':1,'No':0})
```

```python
dataset['Contract'] = dataset['Contract'].replace({'Month-to-month':0,'One year':1,'Two year':2})
```

```python
dataset['MultipleLines'] = dataset['MultipleLines'].replace({'Yes':1,'No':0})
```

```python
dataset.Partner = dataset.Partner.replace({'Yes':1,'No':0})
```

```python
dataset.PaperlessBilling = dataset.PaperlessBilling.replace(ynmap)
```

```python
dataset.Dependents = dataset.Dependents.replace(ynmap)
```
